Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.


Ans :Matplotlib is a widely used plotting library in Python that provides a wide range of high-quality 2D and 3D plotting capabilities. It allows users to create various types of visualizations and graphical representations of data.

Matplotlib is used for data visualization and exploration in fields such as data analysis, scientific research, engineering, finance, and more. It enables users to generate static, animated, or interactive plots, making it a powerful tool for data analysis and communication.

The Pyplot module is a sub-library within Matplotlib that provides a simple interface for creating plots. Here are five types of plots that can be created using the Pyplot module:

Line Plot: It represents data points connected by straight line segments, commonly used for visualizing trends and relationships between variables.

Bar Plot: It displays categorical data with rectangular bars of different heights or lengths, often used to compare and represent discrete data.

Scatter Plot: It shows the relationship between two variables as a collection of individual data points, with one variable on the x-axis and the other on the y-axis. Scatter plots are useful for detecting patterns or correlations in data.

Histogram: It provides a graphical representation of the distribution of a dataset, showing the frequencies of data values within specified intervals (bins). Histograms help in understanding the underlying data distribution.



Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.


A scatter plot is a type of plot that displays the relationship between two variables as a collection of individual data points. It is particularly useful for visualizing the correlation or pattern between two continuous or numerical variables.

In a scatter plot, each data point is represented by a marker on a Cartesian coordinate system. One variable is assigned to the x-axis, and the other variable is assigned to the y-axis. The position of each data point on the plot corresponds to the values of the variables for that particular data point.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

Ans:The subplot() function in Matplotlib is used to create multiple plots or subplots within a single figure. It allows you to arrange and display multiple plots in a grid-like structure, making it easier to compare and analyze different visualizations simultaneously.

By dividing the figure into multiple subplots, you can plot different datasets or different aspects of the same dataset in separate sections of the plot. Each subplot has its own set of axes and can be customized individually.



```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.array([0,1,2,3,4,5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1) 
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()

plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


A box plot, also known as a box-and-whisker plot, is a type of graphical representation that provides a concise summary of the distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers.

A box plot consists of a rectangle (the box) and two lines (the whiskers) extending from the box. The rectangle represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the rectangle represents the median, which divides the data into two equal halves.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.boxplot(data)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.grid(True)

plt.show()
```
